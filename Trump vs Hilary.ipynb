```
#read data
import pandas as pd
import numpy as np
import tensorflow as tf
import os
import math

hilary = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/Trump v.s. Hilary/data/HillaryClinton2014-01-01To2016-10-14Tweets.csv")
trump = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/Trump v.s. Hilary/data/DonaldTrump2014-01-01To2016-10-14Tweets.csv")
```

```
twitter_data = pd.concat([trump,hilary])
twitter_data.head(10)
```

```
twitter_data = twitter_data.loc[twitter_data['retweet'] == False]
```

```
twitter_data = twitter_data.drop(columns=['Unnamed: 0','id','retweet','link'])
```

```
twitter_data = twitter_data.replace({'DonaldTrump':0,'HillaryClinton':1})
```

```
import re

twitter_data['text'] = twitter_data['text'].apply(lambda text: re.sub(r"(?:http|pic.twitter)\S+", "", text))
```

```
twitter_data['text'] = twitter_data['text'].apply(lambda text: re.sub("\n"," ",text))
```

```
twitter_data
```

```
import keras
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(nb_words=3000, lower=True,split=' ')
tokenizer.fit_on_texts(twitter_data['text'].values)
X = tokenizer.texts_to_sequences(twitter_data['text'].values)
X = pad_sequences(X)

Y = twitter_data['author'].values
print(X)
print(Y)
```

```
print(X.shape)
print(Y.shape)
```

```
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense

model = Sequential()
model.add(Embedding(3000,128,input_length=X.shape[1],dropout=0.2))
model.add(LSTM(200, dropout_U = 0.2, dropout_W = 0.2))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer='adam',metrics = ['accuracy'])
model.summary()
```

```
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
model.fit(X_train,y_train,epochs=32,shuffle=True,verbose=1,validation_data=(X_test, y_test))
```

```
trump_tweet = 'I wish people would read or listen to my words on the Border Wall.'
hilary_tweet = 'To all the little girls watching...never doubt that you are valuable and powerful & deserving of every chance & opportunity in the world.'
sample = tokenizer.texts_to_sequences([trump_tweet,hilary_tweet])
sample = pad_sequences(sample,maxlen=32)
print(sample)
```

```
prediction = model.predict_classes(sample)
```
